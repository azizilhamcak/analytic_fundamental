**Tugas Nomor 1 Konversi 6 Jenis Konversi**


    Program Konversi Suhu dengan 6 pilihan menu konversi

```python
def input_menu():
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    print("3. Celcius to Kelvin")
    print("4. Kelvin to Celcius")
    print("5. Fahrenheit to Kelvin")
    print("6. Kelvin to Fahrenheit")
    print("7. Exit\n")


def input_temperature():
      while True:
        try:
            value = float(input("Enter the temperature in Celsius: "))
            return value
        except ValueError:
            print("Please enter a valid number.")

  
def temperatur_conversion(temp):
    input_menu()
    while True:
        try:
            user_input = int(input("Input your choice: "))
            if user_input in [1, 2, 3, 4, 5, 6, 7]:
                if user_input == 1:
                    conv = (9/5*temp)+32
                    print(f"{temp} °C is {conv} °F")
                elif user_input == 2:
                    conv = ((5/9*temp)-32)
                    print(f"{temp} °F is {conv:.2f} °C")
                elif user_input == 3:
                    conv = temp+273.15
                    print(f"{temp} °C is {conv:.2f} °K")
                elif user_input == 4:
                    conv = temp-273.15
                    print(f"{temp} °K is {conv:.2f} °C")
                elif user_input == 5:
                    conv = (5/9*(temp-32))+273.15
                    print(f"{temp} °F is {conv:.2f} °K")
                elif user_input == 6:
                    conv = (9/5*(temp-32))+273.15
                    print(f"{temp} °K is {conv:.2f} °F")
                elif user_input == 7:
                    break
            else:
                print("Input not valid, please input number")
        except ValueError:
            print("Input not valid, please input number")


if __name__ == "__main__":
    temp = input_temperature()
    temperatur_conversion(temp)
```

Dari diagram diatas terdapat outlier harga laptop yang cukup banyak, namun menurut saya hal itu wajar karena dengan harga laptop yang tinggi dengan jumlah permintaan yang sedikit dan diperuntukkan untuk golongan tertentu, ditambah lagi dengan spesifikasi laptop yang sangat berkelas maka data outiler ini tidak berpengaruh. 

Namun, saya ingin mencoba untuk melihat bagaimana jika data outier tadi dihilangkan dan ada berapa banyak data yang berada di luar batas 

```python
# Menghitung IQR
Q1 = df['Price (Euro)'].quantile(0.25)
Q3 = df['Price (Euro)'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas atas dan batas bawah untuk outliers
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

# Menampilkan data yang berada di luar batas
df[(df['Price (Euro)'] < lower_bound) | (df['Price (Euro)'] > upper_bound)]
print(Q1)
print(Q3)
print(lower_bound, upper_bound)
```

```python

dataset_no_outlier = df[(df['Price (Euro)'] > lower_bound) & (df['Price (Euro)'] > upper_bound)] #tanda perbandingan saya buat lebih besar semua untuk melihat data outlier
```
